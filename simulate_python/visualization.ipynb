```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./csv/8.csv')
df['acc_mag'] = np.sqrt(df['acc_x']**2 + df['acc_y']**2 + df['acc_z']**2)
df['gyro_mag'] = np.sqrt(df['gyro_x']**2 + df['gyro_y']**2 + df['gyro_z']**2)

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(8, 12))

# Plot 1: X component
df.plot(x='t', y='acc_mag', ax=ax1, color='red', title='Acceleration')
ax1.set_ylabel('acc_mag')
ax1.grid(True, alpha=0.3)

# Plot 2: Y component  
df.plot(x='t', y='gyro_mag', ax=ax2, color='green', title='Gyro')
ax2.set_ylabel('gyro_mag')
ax2.grid(True, alpha=0.3)

# Plot 3: Z component
df.plot(x='t', y='tof', ax=ax3, color='blue', title='ToF')
ax3.set_ylabel('tof')
ax3.set_xlabel('Time (t)')
ax3.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
