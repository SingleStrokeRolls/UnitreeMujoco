```python
import numpy as np
import os
import pandas as pd
import tensorflow as tf
```

```python
X = []
Y = []
WINDOW_SIZE = 10
for i in range(2000):
    df = pd.read_csv('./csv/'+str(i)+'.csv')
    df = df.iloc[:210,:]
    df['acc_mag'] = np.sqrt(df['acc_x']**2 + df['acc_y']**2 + df['acc_z']**2)
    df['gyro_mag'] = np.sqrt(df['gyro_x']**2 + df['gyro_y']**2 + df['gyro_z']**2)
    for i in range(200):
        data = df.iloc[i:i+WINDOW_SIZE, -3:]
        X.append(data)
        if df.iloc[i,0] > 0.2:
            Y.append(1)
        else:
            Y.append(0)
X = np.array(X)
Y = np.array(Y)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.LSTM(16, use_bias=False, return_sequences=False,
                         input_shape=(10,3)),
    tf.keras.layers.Dense(2, activation='softmax', use_bias=False)
])
model.compile('adam','sparse_categorical_crossentropy',metrics=['acc'])
model.fit(X,Y,batch_size=256,epochs=10,verbose=0)

model.save("0210.keras")
```
